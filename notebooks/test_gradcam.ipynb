```python

import numpy as np
import cv2
import tensorflow as tf

from src.explainability.gradcam import make_gradcam_heatmap, overlay_heatmap

model = tf.keras.models.load_model("saved_models/chest_xray_model.keras")

img_path = "data/raw/nih_sample/images/00000013_005.png"

img = cv2.imread(img_path)
img = cv2.resize(img, (224,224))

img_input = img / 255.0
img_array = np.expand_dims(img_input, axis=0)

heatmap = make_gradcam_heatmap(
    img_array,
    model,
    "conv5_block16_concat"
)

result = overlay_heatmap(heatmap, img)

cv2.imwrite("gradcam_result.png", result)
```
